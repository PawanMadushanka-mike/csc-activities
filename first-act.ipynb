```python
df.rename(columns={'Name':'NAME'},inplace=True)
df.head()
```

```python
#derive a new column
df['new_Backlogs']=round(df['Backlogs']/7,2)
df.head()
```

```python
#delete a column from df
del df['Gender']
df.head()
```

```python
#simple filter condition
filter_df=df[ df['Age']>20 ]
filter_df
```

```python
#filter condition in panda
ne='Research','Higher Studies'
filter_df=df[ (df['Age']>20) | df['Alumni Path'].isin(ne) ]
filter_df
```

```python
df.sort_values(by=['NAME'],ascending=[False],inplace=True)
df
```

```python
#arrange dataframe in ascending or decending order
df.sort_values(by=['NAME','Average GPA'],ascending=[False,True],inplace=True)
df
```

```python
#group by in pandas
df2=df.groupby(['Age']).sum()
df2
```

```python
#group by in pandas
df_new=df.groupby(['Age','Backlogs']).sum()
df_new
df_new.to_csv('/kaggle/working/group_new.csv',sep=',',encoding='utf-8')
```

```python
result=df.groupby('Age').aggregate({'Backlogs':'count','Average GPA':['min','max']})
result
```
